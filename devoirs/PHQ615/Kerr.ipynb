## Calcul de la connexion affine et du tenseur de courbure à partir de la métrique

## Métrique de Kerr

```python
from sympy import *
from sympy import Array

init_printing()

t, r, θ, ϕ, ρρ, Δ, M, a = symbols('t, r, θ, ϕ, ρρ, Δ, M, a')
Δ = r**2 + a**2 -2*M*r
ρρ = r**2 + a**2*cos(θ)**2
gd = Array([
    [1-2*M*r/ρρ, 0, 0, 2*M*r*a*sin(θ)**2/ρρ],
    [0, -ρρ/Δ, 0, 0], 
    [0, 0, -ρρ, 0],
    [2*M*r*a*sin(θ)**2/ρρ, 0, 0, -(r*r+a*a+2*a*a*M*r*sin(θ)**2/ρρ)*sin(θ)**2]
])
coords = Array([t, r,θ, ϕ])
gu = Array([
    [(r**2+a**2)/Δ+2*a**2*M*r*sin(θ)**2/(ρρ*Δ), 0, 0, 2*M*r*a/(ρρ*Δ)],
    [0, -Δ/ρρ, 0, 0], 
    [0, 0, -1/ρρ, 0],
    [2*M*r*a/(ρρ*Δ), 0, 0, (-ρρ + 2*M*r)/(ρρ*Δ*sin(θ)**2)]
])
gd
```

```python
gu
```

```python
simplify(tensorcontraction(tensorproduct(gd, gu), (1,2)))
```

```python
# connexion affine 'covariante'
dg = derive_by_array(gd, coords)
gammad = (permutedims(dg,(2,0,1)) + permutedims(dg,(2,1,0))-dg)/2
gammad = factor(gammad)
gammad
```

```python
# connexion affine
Γ = tensorcontraction(tensorproduct(gu, gammad), (1,2))
Γ = simplify(Γ)
Γ
```

## Calcul du tenseur de courbure

```python
R1 = derive_by_array(Γ, coords)
C = tensorcontraction(tensorproduct(Γ, Γ), (0,4))
R = permutedims(R1,(1,2,0,3)) - permutedims(R1,(1,2,3,0)) + permutedims(C,(2,0,3,1)) - permutedims(C,(2,0,1,3))
R = simplify(R)
R
```

```python
# tenseur de Ricci
Ricci = tensorcontraction(R, (0,2))
Ricci = simplify(Ricci)
Ricci
```
